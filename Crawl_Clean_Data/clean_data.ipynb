```python
import pandas as pd
import re
from datetime import datetime, timedelta
import numpy as np

# Ví dụ: Giả sử df đã có sẵn cột Start_time, End_time, total_time
# df = pd.read_csv('your_file.csv')

# 1. Tách take_place từ Start_time
df[['Start_time_clean', 'take_place']] = df['Start_time'].str.split(r'\n', n=1, expand=True)
df['Start_time_clean'] = df['Start_time_clean'].str.strip()

# 2. Tách destination từ End_time
df[['End_time_clean', 'destination']] = df['End_time'].str.split(r'\n', n=1, expand=True)
df['End_time_clean'] = df['End_time_clean'].str.strip()

# 3. Tạo cột transit từ total_time (phần sau \n), và giữ lại thời gian bay
df[['flight_time', 'transit']] = df['Total_time'].str.split(r'\n', n=1, expand=True)
df['flight_time'] = df['flight_time'].str.strip()

# 4. Chuyển flight_time (vd: "1h 20m") thành số giờ float
def convert_to_hours(time_str):
    if pd.isna(time_str):
        return 0
    h = re.search(r'(\d+)h', time_str)
    m = re.search(r'(\d+)m', time_str)
    hours = int(h.group(1)) if h else 0
    minutes = int(m.group(1)) if m else 0
    return round(hours + minutes / 60, 2)

df['total_time_hour'] = df['flight_time'].apply(convert_to_hours)

# 5. Tạo cột End_day: nếu Start_time + total_time_hour > 24 thì cộng thêm 1 ngày
def calculate_end_day(row):
    try:
        full_start = datetime.strptime(f"{row['Start_Day']} {row['Start_time_clean']}", '%d-%m-%Y %H:%M')
        landing_time = full_start + timedelta(hours=row['total_time_hour'])
        return landing_time.strftime('%d-%m-%Y')
    except Exception as e:
        print(f"Lỗi tại dòng: {row.name}, error: {e}")
        return np.nan

df['End_day'] = df.apply(calculate_end_day, axis=1)
print(df)

# Xóa các cột gốc nếu muốn
df = df.drop(['Start_time', 'End_time', 'Total_time'], axis=1)

# Đổi tên lại cho dễ hiểu
df.rename(columns={
    'Start_time_clean': 'Start_time',
    'End_time_clean': 'End_time'
}, inplace=True)

```

```python
import pandas as pd

# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv(r'D:\ADMIN\Desktop\crawlingdata\tour_riel.csv')
print(df.info())
# Display the first few rows of the dataframe
display(df)
```

```python
df = df[~df['price'].isna()]

print(df['price'].isna().sum())
# Lọc các dòng mà cột Price không chứa chữ số (0-9)
invalid_price_rows = df[~df['price'].str.contains(r'\d', na=False)]

# In ra các dòng không có số trong Price
print(invalid_price_rows)
df = df[df['price'].str.contains(r'\d', na=False)].reset_index(drop=True)
```

```python
# Bước 1: Loại bỏ dấu chấm và phần "VND"
df['price'] = df['price'].replace({r'\.': '', r' VND': ''}, regex=True)

# Bước 2: Chuyển đổi giá trị sang kiểu số (int)
df['price'] = df['price'].astype(int)

import re

def convert_duration_to_hours(value):
    try:
        if pd.isna(value):
            return None
        value = value.strip()

        # Tìm số giờ và số phút nếu có
        hours_match = re.search(r'(\d+)\s*Hour', value)
        minutes_match = re.search(r'(\d+)\s*Minute', value)

        hours = int(hours_match.group(1)) if hours_match else 0
        minutes = int(minutes_match.group(1)) if minutes_match else 0

        total_hours = round(hours + minutes / 60, 2)
        return total_hours
    except Exception as e:
        print(f"Lỗi: {e} với giá trị: {value}")
        return None

# Áp dụng
df['duration'] = df['duration'].apply(convert_duration_to_hours)
df['location'] = df['location'].str.replace('\n', ' ', regex=True).str.strip()

display(df)
```

```python
df.to_csv('tour_clean.csv', index=False)
```

```python
import pandas as pd

# Đọc dữ liệu
df = pd.read_csv('hotel_clean_new.csv')

# Khởi tạo cột start_clean = star_number
df['start_clean'] = df['star_number']

# Danh sách xử lý theo thành phố
city_groups = {
    'Đà Nẵng': [0.60, 0.90],
    'Phú Quốc': [0.60, 0.90],
    'Đà Lạt': [0.05, 0.20, 0.70],
    'Thành Phố Hạ Long': [0.30, 0.60]
}

# Hàm phân sao theo thành phố
def assign_star_by_city(price, city, quantiles):
    if city in ['Đà Nẵng', 'Phú Quốc']:
        if price > quantiles[0.90]:
            return 5
        elif price > quantiles[0.60]:
            return 4
        else:
            return 3
    elif city == 'Đà Lạt':
        if price > quantiles[0.70]:
            return 5
        elif price > quantiles[0.20]:
            return 4
        elif price > quantiles[0.05]:
            return 3
        else:
            return 2
    elif city == 'Thành Phố Hạ Long':
        if price > quantiles[0.60]:
            return 5
        elif price > quantiles[0.30]:
            return 4
        else:
            return 3
    else:
        return None  # Không thay đổi

# Áp dụng logic cho từng thành phố
for city, qs in city_groups.items():
    # Chọn dữ liệu phù hợp
    cond = (df['hotel_city'] == city)
    subset = df.loc[cond]
    
    # Tính phân vị cho giá tại thành phố đó
    quantiles = subset['price'].quantile(qs)

    # Gán lại start_clean
    df.loc[cond, 'start_clean'] = subset.apply(
        lambda row: assign_star_by_city(row['price'], city, quantiles), axis=1
    )

# In kết quả
print(df[['hotel_name', 'hotel_city', 'price', 'star_number', 'start_clean']])
```

```python
df.to_csv('hotel_clean_new.csv', index=False)
```
